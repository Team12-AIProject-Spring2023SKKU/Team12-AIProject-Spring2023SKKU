```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

import numpy as np
import pandas as pd
```

```python
for state in ['AR', 'CA', 'IA', 'IL', 'LA', 'MN', 'ND', 'NE', 'SD']:
    df_state = df[df['state'] == state]

    for month in range(1, 13):
        df_month = df_state[df_state['ID'].str.contains('-{:02}'.format(month))]

        for colname in ['productions_spring', 'harvested_spring', 'cultivated_spring']:
            df_col = df_month[colname]

            train = df_col[df_col.notnull()][:7]
            print()
            print(state, month, colname)

            train_X = train.index
            train_X = np.array(train_X).reshape(-1, 1)
            line_model = LinearRegression()

            train_y = train

            reg = line_model.fit(train_X, train_y)

            print('ti:', train.index[0])
            print(len(df_col.index))
            pred = df_col[df_col.index < train.index[0]][df_col.isnull()]

            pred_X = pred.index
            pred_X = np.array(pred_X).reshape(-1, 1)
            pred_y = reg.predict(pred_X)

            if np.min(pred_y) < 0:
                pred_y[:] = abs(np.median(pred_y))

            df.loc[pred.index, colname] = pred_y

            for i in range(pred_y.size):
                print(pred_X[i][0], pred_y[i])
```

```python
# First
for colname in ['population_us_asian', 'population_world', 'population_asia_all', 'population_asia_east&southeast', 'yield_rice_asia(hg/ha)', 'production_rice_asia(ton)']:
    df_col = df[colname]

    train = df_col[df_col.notnull()][:2]
    print()
    print(state, month, colname)

    train_X = train.index
    train_X = np.array(train_X).reshape(-1, 1)
    line_model = LinearRegression()

    train_y = train

    reg = line_model.fit(train_X, train_y)

    print('ti:', train.index[0])
    print(len(df_col.index))
    pred = df_col[df_col.index < train.index[0]][df_col.isnull()]

    pred_X = pred.index
    pred_X = np.array(pred_X).reshape(-1, 1)
    try:
        pred_y = reg.predict(pred_X)

        if np.min(pred_y) < 0:
            pred_y[:] = abs(np.median(pred_y))

        df.loc[pred.index, colname] = pred_y

        for i in range(pred_y.size):
            print(pred_X[i][0], pred_y[i])
    except ValueError:
        print('pass')
```

```python
# Last
for colname in ['population_us_asian', 'population_world', 'population_asia_all', 'population_asia_east&southeast', 'yield_rice_asia(hg/ha)', 'production_rice_asia(ton)']:
    df_col = df[colname]

    train = df_col[df_col.notnull()][-2:]
    print()
    print(state, month, colname)

    train_X = train.index
    train_X = np.array(train_X).reshape(-1, 1)
    line_model = LinearRegression()

    train_y = train

    reg = line_model.fit(train_X, train_y)

    print('ti:', train.index[0])
    print(len(df_col.index))
    pred = df_col[df_col.index > train.index[0]][df_col.isnull()]

    pred_X = pred.index
    pred_X = np.array(pred_X).reshape(-1, 1)
    try:
        pred_y = reg.predict(pred_X)

        if np.min(pred_y) < 0:
            pred_y[:] = abs(np.median(pred_y))

        df.loc[pred.index, colname] = pred_y

        for i in range(pred_y.size):
            print(pred_X[i][0], pred_y[i])
    except ValueError:
        print('pass')
```

```python
# Middle
for colname in ['population_us_asian', 'population_world', 'population_asia_all', 'population_asia_east&southeast', 'yield_rice_asia(hg/ha)', 'production_rice_asia(ton)']:
    df_col = df[colname]

    while True:
        null_idx = df_col[df_col.isnull()].index
        if null_idx.size == 0:
            break
        
        pivot = null_idx[0]
        df_left = df_col[df_col.index < pivot]
        start = df_left[df_left.notnull()].index[-1]
        df_right = df_col[df_col.index > pivot]
        end = df_right[df_right.notnull()].index[0]

        train = df_col.iloc[start:end+1][df_col.notnull()]
        print()
        print(state, month, colname)

        train_X = train.index
        train_X = np.array(train_X).reshape(-1, 1)
        line_model = LinearRegression()

        train_y = train

        reg = line_model.fit(train_X, train_y)

        print('ti:', train.index[0])
        print(len(df_col.index))
        pred = df_col.iloc[start:end+1][df_col.isnull()]

        pred_X = pred.index
        pred_X = np.array(pred_X).reshape(-1, 1)
        try:
            pred_y = reg.predict(pred_X)

            if np.min(pred_y) < 0:
                pred_y[:] = abs(np.median(pred_y))

            df.loc[pred.index, colname] = pred_y

            for i in range(pred_y.size):
                print(pred_X[i][0], pred_y[i])
        except ValueError:
            print('pass')
```

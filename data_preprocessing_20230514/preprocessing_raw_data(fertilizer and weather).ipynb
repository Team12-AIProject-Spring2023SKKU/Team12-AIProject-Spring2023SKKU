```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

import numpy as np
import pandas as pd
```

```python
fert = pd.read_csv('./fertilizer_raw.csv')
fert
```

```python
fert_actual = fert[fert['type']=='actual']
print(fert_actual)
fert_predicted = fert[fert['type']=='predicted']
print(fert_predicted)
```

```python
for colname in ['fertilizer_price_index_all', 'fertilizer_price_index_nitrogen', 'fertilizer_price_index_phosphate']:
    train_X = fert_actual.index
    train_X = np.array(train_X).reshape(-1, 1)

    poly = PolynomialFeatures(degree=1, include_bias=True)
    train_poly = poly.fit_transform(train_X)

    line_model = LinearRegression()

    train_y = fert_actual[colname]

    reg = line_model.fit(train_poly, train_y)

    print('train acc:', reg.score(train_poly, train_y))

    pred_X = fert_predicted.index
    pred_X = np.array(pred_X).reshape(-1, 1)
    pred_poly = poly.fit_transform(pred_X)
    pred_y = reg.predict(pred_poly)

    for i in range(1120, len(pred_y)):
        print(pred_y[i])
```

```python
for colname in ['fertilizer_price_index_all', 'fertilizer_price_index_nitrogen', 'fertilizer_price_index_phosphate']:
    train_X = fert_actual[fert_actual.index < fert_actual.index.min() + 14].index
    train_X = np.array(train_X).reshape(-1, 1)

    poly = PolynomialFeatures(degree=1, include_bias=True)
    train_poly = poly.fit_transform(train_X)

    line_model = LinearRegression()

    train_y = fert_actual[fert_actual.index < fert_actual.index.min() + 14][colname]

    reg = line_model.fit(train_poly, train_y)

    print('train acc:', reg.score(train_poly, train_y))

    pred_X = fert_predicted.index
    pred_X = np.array(pred_X).reshape(-1, 1)
    pred_poly = poly.fit_transform(pred_X)
    pred_y = reg.predict(pred_poly)

    for i in range(1140):
        print(pred_y[i])
```

Regression for the weather data

```python
weather = pd.read_csv('./weather_raw.csv')
weather
```

```python
weather[weather['state'] == 'AR']
```

```python
weather.loc[[1, 2]]['ID']
```

```python
for state in ['AR', 'CA', 'IA', 'IL', 'LA', 'MN', 'ND', 'NE', 'SD']:
    weather_state = weather[weather['state'] == state]

    for month in range(1, 13):
        weather_month = weather_state[weather_state['ID'].str.contains('-{:02}'.format(month))]

        for colname in ['temperature_avg', 'CDD', 'GDD', 'HDD', 'temperature_max', 'temperature_min', 'precipitation', 'snow_depth', 'snow_fall']:
            weather_col = weather_month[colname]

            train = weather_col[weather_col.notnull()]
            print()
            print(state, month, colname)

            train_X = train.index
            train_X = np.array(train_X).reshape(-1, 1)
            line_model = LinearRegression()

            train_y = train

            reg = line_model.fit(train_X, train_y)

            pred = weather_col[weather_col.isnull()]

            pred_X = pred.index
            pred_X = np.array(pred_X).reshape(-1, 1)
            pred_y = reg.predict(pred_X)

            weather.loc[pred.index, colname] = pred_y

            for i in range(pred_y.size):
                print(pred_X[i][0], pred_y[i])
```

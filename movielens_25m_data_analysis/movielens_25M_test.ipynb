```python
lowest = user_movie_rating[user_movie_rating['rating'] == 0.5]
lowest
```

```python
for genre in ['Drama', 'Comedy', 'Thriller', 'Romance', 'Action', 'Horror', 'Documentary', 'Crime', 'Adventure', 'Sci-Fi', 'Children', 'Animation', 'Mystery', 'Fantasy', 'War', 'Western', 'Musical', 'Film', 'Noir', 'IMAX']:
    p = lowest[lowest['genres'].str.contains(genre)]
    count = p.shape[0]

    
    a = user_movie_rating[user_movie_rating['genres'].str.contains(genre)]
    call = a.shape[0]

    print("{} {} {}".format(genre, count, call))
```
